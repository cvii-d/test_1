```python
#imports 
import numpy as np #for array handling
import pandas as pd 
from pandas import Series, DataFrame 
```

```python
#we need data to start working
import webbrowser
website = 'http://en.wikipedia.org/wiki/NFL_win-loss_records'
webbrowser.open(website) # opening the website 
```

```python
#copy and paste it in pandas using the clipboard - you need to click on copy the text
nfl_frame = pd.read_clipboard()
```

```python
#now pandas will read whats on your clipboard 
nfl_frame
```

```python
#grabbign column names 
nfl_frame.columns
```

```python
#find column values
nfl_frame.Team
```

```python
#grabbing columns with more names
nfl_frame['First NFL season']
```

```python
#grabbing multiple columns
DataFrame(nfl_frame,columns=['Team','First NFL season','Division'])
```

```python
#asking for non existing column
DataFrame(nfl_frame,columns=['Team','First NFL season','Stadium'])
```

```python
#retrive rows from the dataframe
nfl_frame
```

```python
#retrive first few rows in the brackets (5)
nfl_frame.head(3)
```

```python
#retrives the last few rows
nfl_frame.tail(3)
```

```python
#retrive rows on particular index - .ix is now .iloc
nfl_frame.iloc[3]
```

```python
#assign values to entire columns - it fills in the entire column with the value
nfl_frame['Stadium'] = "Levi's Stadium"
```

```python
nfl_frame
```

```python
#pass numbers into the column
nfl_frame['Stadium'] = np.arange(5)
```

```python
nfl_frame
```

```python
#add series to dataframe - put the values in the given index
stadiums = Series(["Levi's Stadium", "AT&T Stadium"],index=[4,0])
stadiums
```

```python
#imputing the values in the dataframe
nfl_frame['Stadium'] = stadiums
```

```python
#already assigned to the columns and position 
nfl_frame
```

```python
#delete entire column
del nfl_frame['Stadium']
```

```python
nfl_frame
```
