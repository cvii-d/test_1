```python
#standart imports
import numpy as np
import pandas as pd
from numpy.random import randn

#stats 
from scipy import stats

#plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#create random normal distribution
dataset1 = randn(100)
```

```python
#plot histogram on the dataset
#plt.hist will always give you 10 bins no matter how many values 
plt.hist(dataset1)
```

```python
dataset2 = randn(80)

plt.hist(dataset2,color='indianred')
```

```python
# We can use normed to plot on same plot

# Set density=True for the plots to be normalized in order to comapre data sets with different number of observations
# Set alpha=0.5 for transperancy
#density is normed 

plt.hist(dataset1,density=True,color='indianred',alpha=0.5,bins=20)
plt.hist(dataset2,density=True,alpha=0.5,bins=20)
```

```python
data1 = randn(1000)
data2 = randn(1000)
```

```python
#joined distribution with joinplot
sns.jointplot(data1,data2)
```

```python
#you can use hex bins
sns.jointplot(data1,data2,kind='hex')
```
