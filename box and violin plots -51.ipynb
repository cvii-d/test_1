```python
#standart imports
import numpy as np
import pandas as pd
from numpy.random import randn

#stats 
from scipy import stats

#plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#create 2 datasets
data1 = randn(100)
data2 = randn(100)
```

```python
#add them to boxplot in a list
sns.boxplot(data=[data1,data2])
```

```python
#outlier points to be added to the whiskers
sns.boxplot(data=[data1,data2],whis=np.infty)
```

```python
#make it horizontally
sns.boxplot(data=data1,orient='h')
```

```python
#violin plots combines simple boxplot and kernel estimation plots
#example where boxplot doenst give the full picture -> create normal distribution
data1 = stats.norm(0,5).rvs(100)

#two gamma distrubitons with concatenated together
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                        -1*stats.gamma(5).rvs(50)])

#boxplot both data1 and data2

sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
#look into the kernel estimation plot to see the full the picture
sns.violinplot(data=[data1,data2])
```

```python
#change the banwith of the ckernel
sns.violinplot(data=[data2],bw=0.01)
```

```python
#include individual points with inner='stick'
sns.violinplot(data=[data1],inner='stick')
```
