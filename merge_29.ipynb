```python
#merge by linking rows by keys
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
#create dataframe
dFrame1 = DataFrame({'key':['X','Z','Y','Z','X','X'], 'data_set1':np.arange(6)})

dFrame1
```

```python
dFrame2 = DataFrame({'key':['Q','Y','Z'],'data_set2':[1,2,3]})

dFrame2
```

```python
#use the merge method - many to one method - choses automatically 
pd.merge(dFrame1,dFrame2)
```

```python
#match on specific column
pd.merge(dFrame1,dFrame2, on='key')
```

```python
#which dataframe keys to use 
#how on the left - merged on the left, and the null respond to missing valus 
pd.merge(dFrame1, dFrame2, on='key',how='left')
```

```python
#for Q matched as it exists on the null 
pd.merge(dFrame1,dFrame2, on ='key', how='right')
```

```python
#now you have all the information on the 2 sets
pd.merge(dFrame1,dFrame2, on='key',how='outer')
```

```python
#many to many merge
dFrame3 = DataFrame({'key':['X','X','X','Y','Z','Z'],'dataset_3':range(6)})
dFrame3
```

```python
dFrame4 = DataFrame({'key':['Y','Y','X','X','Z'], 'dataset_4':range(5)})
dFrame4
```

```python
#merge them
pd.merge(dFrame3,dFrame4)
```

```python
#merge with multiple keys
df_left = DataFrame({'key1':['SF','SF','LA'],
                      'key2':['one','two','one'],
                      'left_data':[10,20,30]})

df_left
```

```python
df_right = DataFrame({'key1':['SF','SF','LA','LA'],
                      'key2':['one','one','one','two'], 
                      'right_data':[40,50,60,70]})
```

```python
df_right
```

```python
#merge on multiple keys
pd. merge(df_left,df_right,on=['key1','key2'],how='outer')
```

```python
#adding suffix
pd.merge(df_left,df_right,on='key1')
```

```python
#if you dont want to have automatical suffix
pd.merge(df_left,df_right,on='key1',suffixes=('_lefty','_righty'))
```
