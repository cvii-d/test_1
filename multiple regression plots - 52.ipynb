```python
#standart imports
import numpy as np
import pandas as pd
from numpy.random import randn

#stats 
from scipy import stats

#plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#upload data set from the seaborn example
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#plot the total bill
sns.lmplot('total_bill','tip',tips)
```

```python
#add dictionnairy to see individual parts of the plot
sns.lmplot('total_bill','tip',tips,
            scatter_kws={'marker':'o','color':'indianred'},
            line_kws={'linewidth':1,'color':'blue'})
```

```python
#multiple higher orders 
#passing order 

sns.lmplot('total_bill','tip',tips, order=4,
            scatter_kws={'marker':'o','color':'indianred'},
            line_kws={'linewidth':1,'color':'blue'})
```

```python
#not fit the regression at all and let it be scatter plot
sns.lmplot('total_bill','tip',tips,fit_reg=False)
```

```python
#percentage of the tip 
#to get it 
tips.head()
```

```python
#create a new column 
tips['tip_perc']=100*(tips['tip']/tips['total_bill'])
```

```python
tips.head()
```

```python
sns.lmplot('size','tip_perc',tips)
```

```python
#to make it easy to read with adding jitter
sns.lmplot('size','tip_perc',tips,x_jitter=.1)
```

```python
#make tendesi of each bin 
sns.lmplot('size','tip_perc',tips,x_estimator=np.mean)
```

```python
# Plot, note the markers argument
sns.lmplot('total_bill','tip_perc',tips,hue='sex',markers=['x','o'])
```

```python
#does the day make the difference for the tip 
sns.lmplot('total_bill','tip_perc',tips,hue='day')
```

```python
#local regression
sns.lmplot('total_bill','tip_perc',tips,lowess=True,line_kws={'color':'black'})
```

```python
#lower level function - regplot
sns.regplot('total_bill','tip_perc',tips,line_kws={'color':'black'})
```

```python
#create a figure of 2 subplots
#colors= is palette=
fig,(axis1,axis2) = plt.subplots(1,2,sharey=True)

sns.regplot('total_bill','tip_perc',tips,ax=axis1)
sns.violinplot(x=tips['size'],y=tips['tip_perc'],palette='Reds_r',ax=axis2)
```
