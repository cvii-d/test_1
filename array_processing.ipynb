```python
#set the grid
points = np.arange(-5,5,0.01)
```

```python
#numpy where

A = np.array([1,2,3,4])

B = np.array([100,200,300,400])
```

```python
#create a boolean array, to imput true or false conditions
condition = np.array([True,True,False,False])
```

```python
#list comprehension, zip to include multipled condition
answer = [(A_val if cond else B_val) for A_val,B_val,cond in zip (A,B,condition)]

#show me a value if the condition is met or show me b value when the condition is met
```

```python
answer
```

```python
#use numpy where is like the comprehension array
answer2 = np.where (condition,A,B)
```

```python
answer2
```
